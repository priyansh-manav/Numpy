```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Data structure: [restaurant_id, 2021, 2022, 2023, 2024]
sales_data = np.array([
    #   2020     2021    2022    2023
    [1, 150000, 180000, 220000, 250000],  # Paradise Biryani
    [2, 120000, 140000, 160000, 190000],  # Beijing Bites
    [3, 200000, 230000, 260000, 300000],  # Pizza Hub
    [4, 180000, 210000, 240000, 270000],  # Burger Point
    [5, 160000, 185000, 205000, 230000]   # Chai Point
])

print("====== Zomato sales analysis =======")
print("Sales Data shape: ",sales_data.shape)
print("Sample data for 1st 3 Restaurant: \n",sales_data[:3, 1:])

# total sales pr year
total_sales = np.sum(sales_data[:,1:],axis=0)
print("\nTotal sales per year: \n",total_sales)

# minimum sales per restaurant
mini_sales_res = np.min(sales_data[:,1:],axis=1)
print("\nMinimum sales per year: \n",mini_sales_res)

# maximum sales per year
max_sales_year = np.max(sales_data[:,1:],axis=0)
print("\nMaximum sales per year: \n",max_sales_year)

# Average sales per restaurant
avg_sales_res = np.mean(sales_data[:,1:],axis=1)
print("\nAverage sales per restaurant: \n",avg_sales_res)

#cumulative_sum
cumsum = np.cumsum(sales_data[:,1:],axis=1)
print("\ncumulative sum: \n",cumsum)

plt.figure(figsize=(8,6))
plt.plot(np.mean(cumsum,axis=0))
plt.title("Average cumulative sales accross all restaurant")
plt.xlabel("year")
plt.ylabel("sales")
plt.grid(True)
plt.show()
```
